```python
X=data.drop(['Teams'],axis=1)
```

```python
plt.figure(figsize=(10, 9))
plt.title("IPL 2024 Dendograms")
plt.xlabel("Teams")
plt.ylabel('Distance')
dend=shc.dendrogram(shc.linkage(X,method='centroid'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=6, linkage='average')
clt1=cluster.fit_predict(X)
```
